```python
df_imp = pd.DataFrame(df_orig)
```

#  Data imputation

```python
df_imp = df_imp.dropna(subset="City")
df_imp
```

```python
df_imp["Temperature(F)"].fillna(df_imp["Temperature(F)"].mean(), inplace=True)
```

```python
proportions_weather = df_imp["Weather_Condition"].value_counts(normalize=True)

def impute_weather_condition(row):
    if pd.isna(row["Weather_Condition"]):
        return np.random.choice(proportions_weather.index, p=proportions_weather)
    else:
        return row["Weather_Condition"]

df_imp["Weather_Condition"] = df_imp.apply(impute_weather_condition, axis=1)
```
